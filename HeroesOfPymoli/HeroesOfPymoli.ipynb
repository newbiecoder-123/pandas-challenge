```python
# Dependencies
import pandas as pd

# File to load 
# Since the notebook is in the location as the csv file, then this is a direct path
file_to_load = "purchase_data.csv"

# Read purchasing file and store into Pandas data frame 
purchase_data = pd.read_csv(file_to_load)
```

```python
purchase_data.head()
```

```python
#Player count, there are duplicate SN's so need to find the total unique players
#value_counts accounts for UNIQUE values while count includes duplicates
player_count = len(purchase_data["SN"].value_counts())

#put player_count into a list, so you don't have to use an index for scalar values
player_count_df = pd.DataFrame({"Total Players": [player_count]})
player_count_df
```

```python
purchase_data.columns
```

```python
#---------------Purchasing Analysis (Total)

#Number of unique items
unique_item = len(purchase_data["Item ID"].unique())
unique_item

#Average purchase price
average_price = round(purchase_data["Price"].mean(),2)
average_price

#Total number of purchases
total_purchase = purchase_data["Purchase ID"].count()
total_purchase

#Total revenue
total_revenue = round(purchase_data["Price"].sum(),2)
total_revenue

purchasing_summary_df = pd.DataFrame({
    "Number of Unique Items": [unique_item],
    "Average Price": [average_price],
    "Number of Purchases": [total_purchase],
    "Total Revenue": [total_revenue]
})

#Formatting the values to show dollar signs and comma
purchasing_summary_df["Average Price"] = purchasing_summary_df["Average Price"].map("${:.2f}".format)
purchasing_summary_df["Total Revenue"] = purchasing_summary_df["Total Revenue"].map("${:.2f}".format)

purchasing_summary_df
```

```python
#-----------Gender Demographics

#Use groupby function to group by gender
gender_df = purchase_data.groupby("Gender")

#use "nunique" to return the number of unique SN names instead of unique which just returns the unique names
unique_gender_df = gender_df["SN"].nunique()

gender_summary_df = pd.DataFrame({
    "Total Count": unique_gender_df,  
    "Percentage of Players": unique_gender_df / player_count
})

gender_summary_df["Percentage of Players"] = gender_summary_df["Percentage of Players"].map('{:,.2%}'.format)
gendersummary_sort_df = gender_summary_df.sort_values(["Total Count"], ascending=False)

gendersummary_sort_df
```

```python
#-----------Purchasing analysis (Gender)

#Purchase count per group
purchase_count_df = gender_df["Price"].count()

#Average purchase price per group
average_price_df = gender_df["Price"].mean()

#Total purchase value per group, summing by each group 
total_purchase_value = gender_df["Price"].sum()

#Average Purchase Total per Person by unique shopper
avg_purchase_total_per_person = total_purchase_value / unique_gender_df

gender_summary_df = pd.DataFrame({
    "Purchase Count": purchase_count_df,  
    "Average Purchase Price": average_price_df,
    "Total Purchase Value": total_purchase_value,
    "Avg Purchase Total per Person": avg_purchase_total_per_person
})

gender_summary_df["Average Purchase Price"] = gender_summary_df["Average Purchase Price"].map('${:,.2f}'.format)
gender_summary_df["Total Purchase Value"] = gender_summary_df["Total Purchase Value"].map('${:,.2f}'.format)
gender_summary_df["Avg Purchase Total per Person"] = gender_summary_df["Avg Purchase Total per Person"].map('${:,.2f}'.format)

gender_summary_df
```

```python
#-----------Purchasing Analysis (Age)

#Establish bins for age groups (need 9 bins) 
bins = [0, 9, 14, 19, 24, 29, 34, 39, 999]

# Create the names for the bins
# There are 8 categories, so there will need to be 9 bins
age_labels = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", ">40"]

# Slice the data and place it into bins
purchase_data["Age Group"] = pd.cut(purchase_data["Age"], bins, labels=age_labels)

# Use Groupby to create an indexed data frame by age group
age_group_df = purchase_data.groupby("Age Group")

# Purchase count per group
age_purchase_count_df = age_group_df["Price"].count()

# Average purchase price per group
age_average_price_df = age_group_df["Price"].mean()

# Total purchase value per group, summing by each group 
age_total_purchase_value = age_group_df["Price"].sum()

# Average Purchase Total per Person by unique shopper
age_avg_purchase_total_per_person = age_total_purchase_value / age_group_count

#Create a summary data frame to hold the results
age_group_summary_df = pd.DataFrame({
    "Purchase Count": age_purchase_count_df,  
    "Average Purchase Price": age_average_price_df,
    "Total Purchase Value": age_total_purchase_value,
    "Avg Purchase Total per Person": age_avg_purchase_total_per_person
})

#Formatting
age_group_summary_df["Average Purchase Price"] = age_group_summary_df["Average Purchase Price"].map('${:,.2f}'.format)
age_group_summary_df["Total Purchase Value"] = age_group_summary_df["Total Purchase Value"].map('${:,.2f}'.format)
age_group_summary_df["Avg Purchase Total per Person"] = age_group_summary_df["Avg Purchase Total per Person"].map('${:,.2f}'.format)

age_group_summary_df
```

```python
#--------Top Spenders

#Use Groupby to create an indexed data frame by spenders
spenders_df = purchase_data.groupby("SN")

# Purchase count per group
spenders_purchase_count_df = spenders_df["Price"].count()
# Check by calling a row spenders_purchase_count_df["Lisosia93"]

# Average purchase price per group
spenders_average_price_df = spenders_df["Price"].mean()
# Check by calling a row spenders_average_price_df["Lisosia93"]

# Total purchase value per group, summing by each group
spenders_total_purchase_value = spenders_df["Price"].sum()
# Check by calling a row spenders_total_purchase_value["Lisosia93"]

#Create a summary data frame to hold the results
spenders_summary_df = pd.DataFrame({
    "Purchase Count": spenders_purchase_count_df,  
    "Average Purchase Price": spenders_average_price_df,
    "Total Purchase Value": spenders_total_purchase_value
})

#Formatting
spenders_summary_df["Average Purchase Price"] = spenders_summary_df["Average Purchase Price"].map('${:,.2f}'.format)
spenders_summary_df["Total Purchase Value"] = spenders_summary_df["Total Purchase Value"].map('${:,.2f}'.format)

top_spenders = spenders_summary_df.sort_values(["Total Purchase Value"], ascending=False)
top_spenders.head()
```

```python
#-----------Most Popular Items

#Retrieve the Item ID, Item Name, and Item Price columns
#use iloc and do not include Age Group
item_df = purchase_data.iloc[:, 4:7]

#Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value
grouped_item_df = item_df.groupby(["Item ID","Item Name"])

grouped_item_purchase_count = grouped_item_df["Item ID"].count()
grouped_item_total_purchase_value = grouped_item_df["Price"].sum()
grouped_item_price = grouped_item_total_purchase_value / grouped_item_purchase_count

#Create a summary data frame to hold the results
items_summary_df = pd.DataFrame({
    "Purchase Count": grouped_item_purchase_count,  
    "Item Price": grouped_item_price,
    "Total Purchase Value": grouped_item_total_purchase_value
})

#Formatting
#items_summary_df["Item"] = items_summary_df["Item"].map('${:,.2f}'.format)
#items_summary_df["Total Purchase Value"] = items_summary_df["Total Purchase Value"].map('${:,.2f}'.format)

most_popular_items = items_summary_df.sort_values(["Purchase Count"], ascending=False).head()
most_popular_formatted = most_popular_items.style.format({"Item Price":"${:,.2f}","Total Purchase Value":"${:,.2f}"})

most_popular_formatted
```
